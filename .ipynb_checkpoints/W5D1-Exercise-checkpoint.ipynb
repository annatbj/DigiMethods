```python
import matplotlib.pyplot as plt
values=[item[0] for item in suicide_Nebraska]
labels=[int(item[4]) for item in suicide_Nebraska]

plt.xlabel("Year")
plt.title("Graph showing the number of deaths by suicide in Nebraska")
plt.ylabel("Number of deaths per 100,000")
plt.xticks(range(0,len(labels)),rotation=90)
plt.plot(values,labels)
plt.show()
```

**Exercise 5.1.1.4** Make your figure self-contained and export it/save it to your machine.

```python
import os
values=[item[0] for item in suicide_Nebraska]
labels=[int(item[4]) for item in suicide_Nebraska]

plt.figure(figsize=(10,5),dpi=100)
plt.xlabel("Year")
plt.title("Graph showing the number of suicides in Nebraska per 100,000")
plt.ylabel("Number of deaths per 100,000")
plt.xticks(range(0,len(labels)),rotation=90)
plt.plot(values,labels)

file_dir=os.getcwd()
plt.savefig(file_dir+"/SuicidesNebraska.png")
```

**Exercise 5.1.1.5** Make a list that is a subset of the data, containing _two_ causes of death that might be interesting to compare for one state.

```python
#"Unintentional injuries" "Nebraska"
double_list=[]
double_list.append(suicide_Nebraska)

unintentional_list=[]
for rows in better_list_of_death:
    if rows[2]=="Unintentional injuries" and rows[3]=="Nebraska":
        unintentional_list.append(rows)
double_list.append(unintentional_list)
```

**Exercise 5.1.1.6** Create a _self-contained_ scatter plot that compares the two causes of death in the state.

```python
import matplotlib.pyplot as plt
values_suicide=[int(item[4]) for item in double_list[0]]
values_unintentional=[int(item[4]) for item in double_list[1]]

plt.xlabel("suicide values")
plt.ylabel("unintentional accidents")
plt.plot(values_suicide,values_unintentional, 'o')
```
